## Тест на влияние ограничения ранга

```python
coefss =[]
for n in [2, 3, 4, 5, 6, 7]:
    coefs = []
    m = n * n
    for rank in [5, 10, 20, 30, 40, 50, 60, 70, 80, 100]:
        smp = Sampler(m, n)
        smp.do_MPS_prepare(max_rank = rank, ort_period = 10)
        d = smp.sample_mps(15000)
        coefs.append(smp.check_sample_coef(d))
    coefss.append(coefs)
```
